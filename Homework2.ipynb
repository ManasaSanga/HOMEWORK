```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
# symbol 'GOOG' is for the period of 01-01-2020 to 01-01-2021.
# start_date=datetime.dt(2020-01-01), end_date=datetime.dt(2021-01-02)
to_start='2020-01-01' #to show the start date
to_end='2021-01-01' #to show the end date
google_stock = pdr.get_data_yahoo('GOOG',to_start,to_end)
google_stock
```

```python
#to save the above data into csv file we use
google_stock.to_csv('data_google_stock.csv')
```

```python
# for reading the file we saved into csv we use
google = pd.read_csv('data_google_stock.csv', header=0, index_col='Date')

google
```

```python
#to plot the closing finance for data finance stock we use
google['Close'].plot(grid=True)
#to plot y-axis we use
plt.ylabel('closing price')
#to name the plot we use
plt.title('Data for google stock 2020', y = 1)
#to show the plot we use
plt.show
```
