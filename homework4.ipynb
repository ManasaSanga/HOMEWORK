```python
"""
TASK-1: To Create a graph to find relationship between the sepal length and width, Describe and interpret the graphs results.
I have obtained the graph by using data.csv file and the results are obtained below: """

fig = iris_data[iris_data.species=='setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
iris_data[iris_data.species=='versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='versicolor',ax=fig)
iris_data[iris_data.species=='virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='virginica', ax=fig)

fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

INTERPRETATION OF GRAPH:

In a scatter plot, dots are used to show the values of two different numerical variables. Each dot's location on the horizontal and vertical axes represents a data point's values. To view relationships between variables, utilise scatter plots. Coming to our answer to the graph the iris dataset is having two clusters with fairly clear separation, as seen in the scatter figure above. Iris setosa, which is orange, is present in one of the clusters, while Iris virginica and Iris versicolor are both present in the other cluster (blue and green colors). Since the points are dispersed in an increasing manner from the left to the right and the Iris Setosa cluster exhibits a modest positive correlation, the sepal width is likely to be somewhat influenced by the length of the sepals. The Iris Setosa also displays an anomaly on the left that has to be looked into: a sepal length with a small sepal width. It can be a mistaken entry or a distinct sepal trend. While the size and length of the sepals of Iris virginica and Iris versicolor are visible. The two iris species, iris virginica and iris versicolor, exhibit a concentration of length and width in a single region, indicating that the raph resultsmajority of them have the same long length and narrow width.

```python
"""
TASK-2: To Create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width. Describe and interpret the graph results.

"""
fig = sns.jointplot(x='sepal_length', y='sepal_width', kind='hex', color='red', data=iris_data)
plt.show()
```

INTERPRETATION OF GRAPH:

Joint plot is of three plots and that three plots together make up a jointplot. One of the three plots depicts a bivariate graph that contrasts the variation of the dependent variable (Y) with the independent variable (X). The bivariate graph's top is horizontally positioned with a second plot that displays the distribution of the independent variable (X). The third plot, which displays the distribution of the dependent variable, is positioned on the right border of the bivariate graph and has its orientation set to vertical (Y). The combination of univariate and bivariate plots in a single figure is highly beneficial. This is so that the bivariate analysis can explore the link between two variables and explain the strength of that association, whereas the univariate analysis concentrates on one variable and describes, summarises, and displays any patterns in your data. In the figure shown above, we created a hex plot using the jointplot() function, which by default places two marginal histograms at the top and right margins of the graph. All of the data points that fall within the specified x and y value ranges are contained in the hexagonal bin, which serves as an interval in the display.
The hexagons are coloured using a colour gradient and the number of points falling into each bin is counted. White bins signify that there is no data, while the deeper colours, which emerge in about 6 places, show that the data points are concentrated there. While the sepal length displayed on top shows that the sepal length of 5.5 and 6.0 appears the most in the sepal length recorded, the marginal distribution on the right, which is represented by a historgram and displays the dependent Y variables, shows that the sepal width of 3 appears the most in the flower sepal width recorded.
